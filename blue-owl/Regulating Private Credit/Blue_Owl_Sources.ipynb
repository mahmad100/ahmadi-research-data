## 2. Curated Bibliography

Organized by category. URLs are provided where the source is web-accessible at
a stable address. Filings are linked to SEC EDGAR.

### Statutes, Court Decisions, and Agency Materials (United States)

- **Investment Company Act of 1940 / Investment Advisers Act of 1940**.
  Statutory basis for BDCs and the vacated Private Fund Advisers Rule.
- **Small Business Investment Incentive Act of 1980** (P.L. 96-477).
  Created the BDC structure.
- **Dodd-Frank Wall Street Reform and Consumer Protection Act (2010)**,
  including Section 619 (Volcker Rule).
- **National Association of Private Fund Managers v. SEC**, No. 23-60471
  (5th Cir., Jun 5, 2024). Slip opinion.
  https://www.ca5.uscourts.gov/opinions/pub/23/23-60471CV0.pdf
- **SEC Private Fund Advisers Rule, Release No. IA-6383** (SEC, adopted Aug 23,
  2023; vacated Jun 5, 2024).
- **Goldman v. Blue Owl Capital Inc. et al.**, Case No. 1:25-cv-10047
  (S.D.N.Y., filed Dec 3, 2025). Securities fraud class action.

### Statutes and Agency Materials (International)

- **EU Directive 2024/927 (AIFMD II)**. Adopted March 2024; entered into force
  April 15, 2024; transposition deadline April 16, 2026. Article 15 loan-
  origination rules; Annex V leverage limits (175% NAV open-ended; 300% NAV
  closed-ended).
  https://eur-lex.europa.eu/eli/dir/2024/927/oj/eng
- **UK FCA, "Asset Management & Alternatives Portfolio Letter"** (Mar 1, 2024).
  Launched the multi-firm review of private market valuation governance.
  https://www.fca.org.uk/publication/correspondence/portfolio-letter-asset-management-alternatives-supervisory-strategy-interim-update.pdf
- **ASIC, REP 820 "Private credit surveillance: retail and wholesale funds"**
  (Nov 5, 2025). Surveillance of 28 funds (~A$29.8B) Oct 2024-Aug 2025.
  https://download.asic.gov.au/media/q42bgduw/rep820-published-5-november-2025.pdf
- **ASIC 25-273MR**, "ASIC announces 2026 enforcement priorities" (Nov 2025).
  Private credit named as a 2026 enforcement priority.
  https://www.asic.gov.au/about-asic/news-centre/find-a-media-release/2025-releases/25-273mr-asic-announces-2026-enforcement-priorities/
