# The Numb Market — Companion Charts

Reproducible code behind the Ahmadi Research blog post **[The Numb Market](https://ahmadiresearch.com/blog/numb-market/)** (unpublished draft).

Three charts. Every number is inline. Every source is hyperlinked at the bottom.

> "If someone with the same data can't get to the same conclusion, the analysis isn't done."  
> — [ahmadiresearch.com/approach](https://ahmadiresearch.com/approach.html)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Brand palette — matches assets/css/style.css on the live site.
BLUE_MAIN  = '#005a96'
BLUE_SOFT  = '#0f80c4'
BLUE_LIGHT = '#8fd1ff'
TITLE_BLUE = '#13597B'
GREY_TEXT  = '#374151'
GREY_GRID  = '#e5e7eb'
GREY_LINE  = '#6b7280'

plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'font.size': 10,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.edgecolor': '#9ca3af',
    'axes.labelcolor': GREY_TEXT,
    'xtick.color': GREY_TEXT,
    'ytick.color': GREY_TEXT,
    'axes.grid': True,
    'grid.color': GREY_GRID,
    'grid.linewidth': 0.6,
    'figure.dpi': 110,
    'savefig.dpi': 180,
})
```

## Chart 1 — The shock-response curve

S&P 500 return following each major Trump policy shock between January 2025 and March 2026. Liberation Day is shown as the cumulative two-session decline (April 3 to 4); every other event is the same-day close. The slope flattens with each successive shock: Liberation Day produced a 10.5% drawdown over two sessions, while the Supreme Court striking down the entire IEEPA tariff regime ten months later produced only a 0.69% rally.

Sources: [CNBC](https://www.cnbc.com/2025/04/03/stock-market-today-live-updates.html) [6]; [Supreme Court opinion](https://www.supremecourt.gov/opinions/25pdf/24-1287_4gcj.pdf) [1]; CBOE; [FactSet](https://insight.factset.com/sp-500-earnings-season-update-may-8-2026) [4].

```python
events = [
    ('Mexico/Canada\n25% tariffs\nFeb 1, 2025', -1.8),
    ('Liberation Day\nreciprocal tariffs\nApr 2-4, 2025', -10.5),
    ('U.S. strike on\nIran nuclear sites\nJun 22, 2025', 0.96),
    ('SCOTUS strikes\nIEEPA tariffs\nFeb 20, 2026', 0.69),
    ('Iran closes\nStrait of Hormuz\nMar 4, 2026', 0.5),
]
labels = [e[0] for e in events]
values = np.array([e[1] for e in events])
colors = [BLUE_MAIN if v < 0 else BLUE_LIGHT for v in values]
edges  = [BLUE_MAIN if v < 0 else BLUE_SOFT for v in values]

fig, ax = plt.subplots(figsize=(10, 4.8))
bars = ax.bar(labels, values, color=colors, edgecolor=edges, linewidth=1.3)
ax.axhline(0, color='#9ca3af', linewidth=0.8)
ax.yaxis.set_major_formatter(mticker.PercentFormatter(decimals=0))
ax.set_ylabel('S&P 500 return')
ax.set_title('The shock-response curve has flattened to zero',
             color=TITLE_BLUE, fontsize=13, weight='bold', pad=14, loc='left')
ax.set_ylim(-12, 3)

for bar, v in zip(bars, values):
    y = v + (0.45 if v >= 0 else -0.7)
    ax.text(bar.get_x() + bar.get_width()/2, y, f'{v:+.1f}%',
            ha='center', va='bottom' if v >= 0 else 'top',
            fontsize=9, color=GREY_TEXT)

plt.setp(ax.get_xticklabels(), fontsize=8.5)
fig.tight_layout()
plt.show()
```

## Chart 2 — The bifurcation

The S&P 500 (left axis) and the University of Michigan's Current Economic Conditions sub-index (right axis), monthly, January 2025 through April 2026. The S&P 500 climbs to an all-time high while Current Conditions falls to a level lower than the start of any U.S. recession in the modern series.

Sources: S&P 500 monthly levels via [FRED](https://fred.stlouisfed.org/series/SP500) and [Yahoo Finance](https://finance.yahoo.com/quote/%5EGSPC); University of Michigan, [Surveys of Consumers](https://www.sca.isr.umich.edu/) [3]; [Bloomberg](https://www.bloomberg.com/news/articles/2026-04-24/us-consumer-sentiment-falls-to-record-low-on-inflation) [12].

_Note: S&P 500 values are monthly levels (end-of-month); UMich is the final monthly print. Values are rounded._

```python
months = ['Jan 25','Feb 25','Mar 25','Apr 25','May 25','Jun 25','Jul 25','Aug 25',
          'Sep 25','Oct 25','Nov 25','Dec 25','Jan 26','Feb 26','Mar 26','Apr 26']
# S&P 500 monthly levels (approximate, end-of-month). Apr 25 reflects the Liberation Day drawdown.
spx   = [5910, 6144, 5800, 5283, 5912, 6275, 6450, 6520,
         6680, 6820, 6940, 6890, 6720, 6909, 7050, 7230]
# UMich Current Economic Conditions, monthly final.
umich = [68.7, 65.7, 63.8, 56.5, 58.4, 63.7, 66.8, 64.2,
         60.6, 58.7, 56.1, 55.4, 60.7, 58.6, 54.4, 47.8]

fig, ax1 = plt.subplots(figsize=(10, 4.8))
ax2 = ax1.twinx()
ax2.grid(False)

l1, = ax1.plot(months, spx, color=BLUE_MAIN, linewidth=2.5,
               marker='o', markersize=4, label='S&P 500 (left)')
l2, = ax2.plot(months, umich, color=TITLE_BLUE, linewidth=2.5,
               marker='s', markersize=4, linestyle='--',
               label='UMich Current Conditions (right)')

ax1.set_ylabel('S&P 500', color=BLUE_MAIN)
ax1.tick_params(axis='y', labelcolor=BLUE_MAIN)
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'{v:,.0f}'))
ax1.set_ylim(4800, 7400)

ax2.set_ylabel('UMich Current Conditions', color=TITLE_BLUE)
ax2.tick_params(axis='y', labelcolor=TITLE_BLUE)
ax2.set_ylim(45, 72)

ax1.set_title('Stocks rise. Sentiment collapses. Same window.',
              color=TITLE_BLUE, fontsize=13, weight='bold', loc='left', pad=14)
ax1.legend(handles=[l1, l2], loc='upper center',
           bbox_to_anchor=(0.5, -0.13), ncol=2, frameon=False)
plt.setp(ax1.get_xticklabels(), rotation=0, fontsize=8.5)
fig.tight_layout()
plt.show()
```

## Chart 3 — The April 2025 yield surge

The 10-year U.S. Treasury yield over the trading days surrounding the tariff pause. The yield swung from a 3.87% intraday low on Monday April 7 to a 4.47% intraday high on Wednesday April 9, a roughly 60 basis point move over three sessions and its largest three-day jump since December 2001. Trump announced the 90-day tariff pause on April 9.

The path below is illustrative: April 7 (3.87%) and April 9 (4.47%) are the intraday extremes, not daily closes. Substitute the daily close series from [FRED DGS10](https://fred.stlouisfed.org/series/DGS10) if you want exact end-of-day values.

Sources: [Bloomberg](https://www.bloomberg.com/news/articles/2025-04-11/us-treasury-selloff-is-worst-since-repo-market-chaos-in-2019) [14]; [NY Fed Liberty Street Economics](https://libertystreeteconomics.newyorkfed.org/2026/04/treasury-market-liquidity-since-april-2025/) [14]; [CNBC](https://www.cnbc.com/2025/04/10/trump-dodged-a-disaster-from-the-bond-market-but-the-damage-isnt-over-yet.html) [15].

```python
labels3 = ['Mar 31','Apr 1','Apr 2','Apr 3','Apr 4',
           'Apr 7','Apr 8','Apr 9','Apr 10','Apr 11','Apr 14']
# Illustrative path; Apr 7 (3.87%) and Apr 9 (4.47%) are intraday extremes, not daily closes.
yields  = [4.20, 4.18, 4.20, 4.05, 3.95, 3.87, 4.18, 4.47, 4.40, 4.35, 4.32]

fig, ax = plt.subplots(figsize=(10, 4.8))
ax.plot(labels3, yields, color=BLUE_MAIN, linewidth=2.5)
ax.fill_between(labels3, yields, min(yields) - 0.05, color=BLUE_MAIN, alpha=0.10)
ax.plot([labels3[5]], [yields[5]], 'o', markersize=10, color=TITLE_BLUE, zorder=5)
ax.plot([labels3[7]], [yields[7]], 'o', markersize=10, color=TITLE_BLUE, zorder=5)

ax.annotate('3.87%\nMonday intraday low',
            xy=(labels3[5], yields[5]), xytext=(-55, -50), textcoords='offset points',
            fontsize=9, ha='center', color=GREY_TEXT,
            arrowprops=dict(arrowstyle='-', color='#9ca3af', lw=0.8))
ax.annotate('4.47%\nWednesday intraday peak\ntariff pause announced',
            xy=(labels3[7], yields[7]), xytext=(30, 12), textcoords='offset points',
            fontsize=9, ha='left', color=GREY_TEXT,
            arrowprops=dict(arrowstyle='-', color='#9ca3af', lw=0.8))

ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'{v:.2f}%'))
ax.set_ylim(3.7, 4.65)
ax.set_ylabel('10-year Treasury yield')
ax.set_title('60 basis points in three sessions. Largest three-day jump since December 2001.',
             color=TITLE_BLUE, fontsize=13, weight='bold', loc='left', pad=14)
fig.tight_layout()
plt.show()
```
